Using MinMax scaler for scaling the  numeric columns as they are of different units

```python
from sklearn.preprocessing import MinMaxScaler
mns= MinMaxScaler()
```

```python

ab_new[['Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight']] = mns.fit_transform(ab_new[['Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight']])

ab_new
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
```

Splitting the data into x & y

```python
x=ab_new.iloc[:,0:-1]
x
```

```python
y=ab_new.iloc[:,-1]
y
```

```python
x.shape
```

```python
y.shape
```

Splitting the data into Testing & Training

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30, random_state=45)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

```python
x_preditc=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
```

```python
print("Mean Absolute error:",mean_absolute_error(y_test,x_preditc))
print("Mean Squared error:", mean_squared_error(y_test,x_preditc))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_preditc)))
print(r2_score(y_test,x_preditc))
```

Checking for overfitting or underfitting of the data with lasso,ridge & elastic net

```python

from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
#Lasso Regulatisation Method
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
print('score:',ls.score(x_train,y_train))
print("Mean Absolute error:",mean_absolute_error(y_test,x_preditc))
print("Mean Squared error:", mean_squared_error(y_test,x_preditc))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_preditc)))
print('The coefficient values:',ls.coef_)
print('The intercept values are:',ls.intercept_)
print(r2_score(y_test,x_preditc))
```

```python
#Ridge Regulatisation Method
rg=Ridge(alpha=0.0001)
rg.fit(x_train,y_train)
print('Score:',rg.score(x_train,y_train))
print("Mean Absolute error:",mean_absolute_error(y_test,x_preditc))
print("Mean Squared error:", mean_squared_error(y_test,x_preditc))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_preditc)))
print('The coefficient values:',rg.coef_)
print('The intercept values are:',rg.intercept_)
print(r2_score(y_test,x_preditc))
```

```python
#ElasticNet Regulatisation Method
en=ElasticNet(alpha=0.0001)
en.fit(x_train,y_train)
print('Score:',en.score(x_train,y_train))
print("Mean Absolute error:",mean_absolute_error(y_test,x_preditc))
print("Mean Squared error:", mean_squared_error(y_test,x_preditc))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_preditc)))
print('The coefficient values:',en.coef_)
print('The intercept values are:',en.intercept_)
print(r2_score(y_test,x_preditc))
```

CV score of Linear Regression

```python
from sklearn.model_selection import cross_val_score
```

Cross Validation score of Linear Regression

```python
scr=cross_val_score(lr,x,y,cv=5)
print(scr)
print(scr.mean())
print(round(scr.std()),2)
```

Cross Validation score of Lasso Regression

```python
scr=cross_val_score(ls,x,y,cv=5)
print(scr)
print(scr.mean())
print(round(scr.std()),2)
```

Cross Validation score of Ridge Regression

```python
scr=cross_val_score(rg,x,y,cv=5)
print(scr)
print(scr.mean())
print(round(scr.std()),2)
```

Cross Validation score of ElasticNet Regression

```python
scr=cross_val_score(en,x,y,cv=5)
print(scr)
print(scr.mean())
print(round(scr.std()),2)
```
