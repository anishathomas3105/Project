Checking for Multi collinearity through VIF

```python
import statsmodels.api as sm
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x.shape
```

```python
#to take out the columns alone
x.shape[1]
```

```python
def cal_vif(x):
        vif=pd.DataFrame()
        vif['Variables']=x.columns
        vif['VIF Factors']=[variance_inflation_factor(x.values,i) for i in range (x.shape[1])]
        return(vif)
```

```python
cal_vif(x)
```

```python
print("Mean Absolute error:",mean_absolute_error(y_test,x_predict))
print("Mean Squared error:", mean_squared_error(y_test,x_predict))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_predict)))
print(r2_score(y_test,x_predict))
```

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
#Lasso Regulatisation Method
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
print('score:',ls.score(x_train,y_train))
print("Mean Absolute error:",mean_absolute_error(y_test,x_predict))
print("Mean Squared error:", mean_squared_error(y_test,x_predict))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_predict)))
print('The coefficient values:',ls.coef_)
print('The intercept values are:',ls.intercept_)
print(r2_score(y_test,x_predict))
```

```python
#Ridge Regulatisation Method
ri=Ridge(alpha=0.0001)
ri.fit(x_train,y_train)
print('score:',ri.score(x_train,y_train))
print("Mean Absolute error:",mean_absolute_error(y_test,x_predict))
print("Mean Squared error:", mean_squared_error(y_test,x_predict))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_predict)))
print('The coefficient values:',ri.coef_)
print('The intercept values are:',ri.intercept_)
print(r2_score(y_test,x_predict))
```

```python
#ElasticNet Regulatisation Method

en=ElasticNet(alpha=0.0001)
en.fit(x_train,y_train)
print('score:',en.score(x_train,y_train))
print("Mean Absolute error:",mean_absolute_error(y_test,x_predict))
print("Mean Squared error:", mean_squared_error(y_test,x_predict))
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,x_predict)))
print('The coefficient values:',en.coef_)
print('The intercept values are:',en.intercept_)
print(r2_score(y_test,x_predict))
```
